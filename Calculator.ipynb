```python
import tkinter as tk
from tkinter import messagebox


def evaluate_expression():
    try:
        result = eval(entry.get())
        result_label.config(text=f"Result: {result}")
    except Exception as e:
        messagebox.showerror("Error", f"Invalid input: {e}")


def reset_calculator():
    entry.delete(0, tk.END)
    result_label.config(text="Result: ")


root = tk.Tk()
root.title("Simple Calculator")


entry = tk.Entry(root, width=20, font=('Arial', 16))
entry.grid(row=0, column=0, columnspan=4)


buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+'
]

def button_click(value):
    current_text = entry.get()
    if value == '=':
        evaluate_expression()
    else:
        entry.delete(0, tk.END)
        entry.insert(tk.END, current_text + value)

row, col = 1, 0
for button in buttons:
    if button == '=':
        tk.Button(root, text=button, command=lambda: button_click('='), width=5, height=2, font=('Arial', 16)).grid(row=row, column=col, padx=5, pady=5)
    else:
        tk.Button(root, text=button, command=lambda b=button: button_click(b), width=5, height=2, font=('Arial', 16)).grid(row=row, column=col, padx=5, pady=5)
    col += 1
    if col > 3:
        col = 0
        row += 1

reset_button = tk.Button(root, text='Reset', command=reset_calculator, width=5, height=2, font=('Arial', 16))
reset_button.grid(row=row, column=0, columnspan=4, padx=5, pady=5)


result_label = tk.Label(root, text="Result: ", font=('Arial', 16))
result_label.grid(row=row + 1, column=0, columnspan=4, padx=5, pady=5)


root.mainloop()
```
